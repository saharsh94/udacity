```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('Untitled.jpg')
print(img.shape)
plt.imshow(img)
```

```python
img_copy = np.copy(img)
img_copy = cv2.cvtColor(img_copy , cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

```python
lower_red = np.array([120,0,0])
upper_red = np.array([255,50,50])
```

```python
mask = cv2.inRange(img_copy,lower_red,upper_red)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask_copy = np.copy(img_copy)
mask_copy[mask !=0] =[0,0,0]
plt.imshow(mask_copy)
```

```python
background = cv2.imread('creepy-dark-park-with-two-people-distance.jpg')
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
crop = background[0:335, 0:435]
crop[mask == 0]=[0, 0, 0]
plt.imshow(crop)
```

```python
total = crop + mask_copy
plt.imshow(total)
```

```python
r = img_copy[:,:,0]
g = img_copy[:,:,1]
b = img_copy[:,:,2]

f, (ax1 , ax2 , ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('red')
ax1.imshow(r, cmap='gray')

ax2.set_title('green')
ax2.imshow(g ,cmap='gray')

ax3.set_title('blue')
ax3.imshow(b , cmap='gray')
```

```python
hsv = cv2.cvtColor(img_copy , cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
```

```python
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1 , ax2 , ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('saurate')
ax2.imshow(s ,cmap='gray')

ax3.set_title('valaue')
ax3.imshow(v , cmap='gray')
```

```python
lower_heu = np.array([160, 0 ,0])
upper_heu = np.array([180, 250 ,250])
mask= cv2.inRange(hsv,lower_heu,upper_heu)
hsv_copy=np.copy(hsv)
hsv_copy[mask == 0]=[0,0,0]
plt.imshow(hsv_copy)
```
